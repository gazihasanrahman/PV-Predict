```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# load the dataset 
data = pd.read_csv("LendingClubLoans2018-2020.csv")
```

```python
# see the data shape 
data.shape
```

```python
# converting percentage columns to numeric columns
data['int_rate'] = pd.to_numeric(data['int_rate'].str.strip('%')).div(100)
data['revol_util'] = pd.to_numeric(data['revol_util'].str.strip('%')).div(100)

# check the result
data.loc[:, ['int_rate', 'revol_util']]
```

```python
# converting date columns to month remaining as of today

# convert string to datetime

data.earliest_cr_line = pd.to_datetime(data.earliest_cr_line, format="%b-%y")
data.next_pymnt_d = pd.to_datetime(data.next_pymnt_d, format="%b-%y")
data.issue_d = pd.to_datetime(data.issue_d, errors = 'coerce')
data.last_pymnt_d = pd.to_datetime(data.last_pymnt_d, errors = 'coerce')

# finding the current date
curr_date = pd.to_datetime('today').normalize()

# replacing null value with current date
data['next_pymnt_d'] = data['next_pymnt_d'].replace(np.nan, curr_date)

# calculateing the remamining month
data.earliest_cr_line = ((data.earliest_cr_line - curr_date)/np.timedelta64(1, 'M')).astype(int)
data.issue_d = ((data.issue_d - curr_date)/np.timedelta64(1, 'M')).astype(int)
data.last_pymnt_d = ((data.last_pymnt_d - curr_date)/np.timedelta64(1, 'M')).astype(int)
data.next_pymnt_d = ((data.next_pymnt_d - curr_date)/np.timedelta64(1, 'M')).astype(int)

# changing back the null values
data.next_pymnt_d = data.next_pymnt_d.where(0>data.next_pymnt_d, np.nan)

# check the result
data.loc[:, ['earliest_cr_line', 'issue_d', 'last_pymnt_d', 'next_pymnt_d']]
```

```python
# see the null values and data types 
data.info(verbose=True, null_counts=True)
```

```python
# list of continuous columns and their description
con_cols = (data.select_dtypes(exclude=['object']).columns)
con_cols = con_cols.drop(['id'])
con_desc = data[con_cols].describe()

for column in con_cols:
    con_desc.loc['median',column] = data[column].median()
    con_desc.loc['missing',column] = data[column].isnull().sum()
con_desc
```

```python
# list of categorical columns and their description
cat_cols = (data.select_dtypes(include=['object']).columns)
cat_desc = data[cat_cols].describe()
cat_desc

for column in cat_cols:
    cat_desc.loc['missing',column] = data[column].isnull().sum()
    cat_desc.loc['values',column] = data[column].unique()
cat_desc
```

```python
# list of columns which have more than 60% missing data
percent_missing = data.isnull().sum() * 100 / len(data)
missing_value_df = pd.DataFrame({'column_name': data.columns,
                                 'percent_missing': percent_missing})
missing_values = missing_value_df.loc[(missing_value_df.percent_missing>60),'column_name'].values
print(missing_values)
```

```python
# split the dataset into features and label
X = data.drop(['loan_status'], axis=1)
y = data['loan_status']
```

```python
# split the dataset using the train-test split to avoid overfitting
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)

# check the data shape now 
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# drop the columns which have more than 60% missing data

X_train = X_train.drop(columns= missing_values)
X_test = X_test.drop(columns= missing_values)

# also drop the colums 'id' as it is irrelevent for prediction
X_train = X_train.drop(columns= ['id'])
X_test = X_test.drop(columns= ['id'])

# check the data shape now 
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# convert the categorical data to numeric data
import category_encoders as ce

cat_cols = (X_train.select_dtypes(include=['object']).columns)
enc = ce.OrdinalEncoder(cols=cat_cols.values, handle_missing='return_nan',return_df= True)
X_train = enc.fit_transform(X_train)
X_test = enc.transform(X_test)

# check the data types now 
X_train.info(verbose=True, null_counts=True)
```

```python
# fill the missing value using KNNImputer 

#import the relevant methods
from sklearn.impute import KNNImputer
nan = np.nan

# save the column names first
X_train_col = X_train.columns
X_test_col = X_test.columns

#Apply KNN imputer
imputer = KNNImputer(n_neighbors=2, weights="uniform")
imputed_X_train = imputer.fit_transform(X_train)
imputed_X_test = imputer.transform(X_test)

# Convert output to the data frames
X_train = pd.DataFrame.from_records(imputed_X_train)
X_train.columns = X_train_col

X_test = pd.DataFrame.from_records(imputed_X_test)
X_test.columns = X_test_col

# check the data types now, there is no null values
X_train.info(verbose=True, null_counts=True)
```

```python
# check if there is any constant in the dataset

from sklearn.feature_selection import VarianceThreshold

# setting threshold to zero
var_thres=VarianceThreshold(threshold=0)
var_thres.fit(X_train)
len(X_train.columns[var_thres.get_support()])

constant_columns = [column for column in X_train.columns
                    if column not in X_train.columns[var_thres.get_support()]]

# remove constant from the dataset
X_train = X_train.drop(constant_columns,axis=1)
X_test = X_test.drop(constant_columns,axis=1)

# print the name of the constant deleted
for column in constant_columns:
    print(column)
```

```python
# Feature Selection based on Mutual Information

from sklearn.feature_selection import mutual_info_classif

# determine the mutual information
mutual_info = mutual_info_classif(X_train, y_train)
mutual_info = pd.Series(mutual_info)
mutual_info.index = X_train.columns
mutual_info.sort_values(ascending=False).head(50)
```

```python
# ploting the ordered mutual_info values per feature
mutual_info.head(50).sort_values(ascending=False).plot.bar(figsize=(20, 8))
```

```python
# we will keep only the top 20 important features

from sklearn.feature_selection import SelectKBest

sel_twenty_cols = SelectKBest(mutual_info_classif, k=20)
sel_twenty_cols.fit(X_train, y_train)
selected_cols = X_train.columns[sel_twenty_cols.get_support()]

X_train = X_train.loc[:, selected_cols]
X_test = X_test.loc[:, selected_cols]
```

```python
# check the correlated features using Pearson Correlation

corrmat = X_train.corr()
fig, ax = plt.subplots()
fig.set_size_inches(11,11)
sns.heatmap(corrmat)
```

```python
# this function will select highly correlated features

def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
# number of features which are more than 80% correlated
corr_features = correlation(X_train, 0.8)
len(set(corr_features))
```

```python
# to balance the dataset we will resample the training data

from sklearn.utils import resample

# merge X_train and y_train 
resampling = X_train.copy()
resampling['loan_status']= y_train.values

# separate minority and majority classes
majority_class_a = resampling[resampling['loan_status']=='Good Loan']
majority_class_b = resampling[resampling['loan_status']=='Current']
minority_class = resampling[resampling['loan_status']=='Bad Loan']

# downnsample majority classes
majority_downsampled_a = resample(majority_class_a, 
                                 replace=True,
                                 n_samples=8665)

majority_downsampled_b = resample(majority_class_b, 
                                 replace=True,
                                 n_samples=8665)

# Combine minority class with downnsampled majority class
downsampled = pd.concat([majority_downsampled_a, majority_downsampled_b, minority_class])



# separte X_train and y_train

X_train= downsampled
X_train = X_train.drop(['loan_status'], axis=1)

y_train = downsampled.loan_status

```

```python
# Apply KNN imputer
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=2, weights="uniform")
X_unsupervised = imputer.fit_transform(X_unsupervised)


# Convert output to the data frames
X_unsupervised = pd.DataFrame.from_records(X_unsupervised)
X_unsupervised.columns = X_unsupervised_col
```

```python
# reduce the dimensions using Principal component analysis

from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
pca.fit(X_unsupervised)
pca_data = pca.transform(X_unsupervised)
X_unsupervised = pd.DataFrame(pca_data, columns=['PC1', 'PC2'])
X_unsupervised.shape
```

```python
# now that we have two dimensional data, we can plot it

plt.figure(figsize=(30,30))
plt.scatter(pca_data[:,0],pca_data[:,1])
plt.xlabel('First principle component')
plt.ylabel('Second principle component')
```

```python
# identify the outliers using DBscan
from sklearn.cluster import DBSCAN

#Use the algorithm for outlier detection, 

outlier_detection = DBSCAN(min_samples = 10, eps = 3)
clusters = outlier_detection.fit_predict(X_unsupervised)

#Count total number of outliers as count of those labelled as -1
TotalOutliers = list(clusters).count(-1)
```

```python
# Unsupervised learning:
# using K-Means to find a suitable number of clusters

from sklearn.cluster import KMeans

kmeans_models = [KMeans(n_clusters=k).fit(X_unsupervised) for k in range (1, 10)]
innertia = [model.inertia_ for model in kmeans_models]

plt.plot(range(1, 10), innertia)
plt.title('Elbow method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Using silhouette score to find a suitable number of clusters

from sklearn.metrics import silhouette_score

silhoutte_scores = [silhouette_score(X_unsupervised, model.labels_) for model in kmeans_models[1:9]]
plt.plot(range(2,10), silhoutte_scores, "bo-")
plt.xticks([2, 3, 4, 5, 6, 7, 8, 9, 10])
plt.title('Silhouette scores vs Number of clusters')
plt.xlabel('Number of clusters')
plt.ylabel('Silhoutte score')
plt.show()
```

```python
# aplying K-Means clustering

km = KMeans(n_clusters = 2)
y_predicted = km.fit_predict(X_unsupervised)
y_predicted
```

```python
# create a dataframe for predicted result
data_predicted = X_unsupervised
data_predicted['cluster'] = y_predicted

# ploting the result of the predicted clusters
plt.figure(figsize=(30,30))
df1 = data_predicted[data_predicted.cluster==0]
df2 = data_predicted[data_predicted.cluster==1]
plt.scatter(df1.PC1,df1['PC2'],color='red')
plt.scatter(df2.PC1,df2['PC2'],color='cyan')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
```
