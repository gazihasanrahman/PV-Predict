```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# load the dataset
data = pd.read_csv("PV_Data.csv")
```

```python
# see the null values and data types 
data.info(verbose=True, null_counts=True)
```

```python
# list of continuous columns and their description
con_cols = (data.select_dtypes(exclude=['object']).columns)
con_desc = data[con_cols].describe()

for column in con_cols:
    con_desc.loc['median',column] = data[column].median()
    con_desc.loc['missing',column] = data[column].isnull().sum()
con_desc
```

```python
# list of categorical columns and their description
cat_cols = (data.select_dtypes(include=['object']).columns)
cat_desc = data[cat_cols].describe()
cat_desc

for column in cat_cols:
    cat_desc.loc['missing',column] = data[column].isnull().sum()
    cat_desc.loc['values',column] = data[column].unique()
cat_desc
```

```python
# list of columns which have more than 50% missing data
percent_missing = data.isnull().sum() * 100 / len(data)
missing_value_df = pd.DataFrame({'column_name': data.columns,
                                 'percent_missing': percent_missing})
missing_values = missing_value_df.loc[(missing_value_df.percent_missing>99),'column_name'].values
print(missing_values)
```

```python
# split the dataset into features and label
X = data.drop(['VA'], axis=1)
y = data['VA']
```

```python
# split the dataset using the train-test split to avoid overfitting
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# check the data shape now 
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# drop the columns which have more than 50% missing data

X_train = X_train.drop(columns= missing_values)
X_test = X_test.drop(columns= missing_values)

# check the data shape now 
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# check the data types now 
X_train.info(verbose=True, null_counts=True)
```

```python
# convert the categorical data to numeric data
import category_encoders as ce

cat_cols = (X_train.select_dtypes(include=['object']).columns)
enc = ce.OrdinalEncoder(cols=cat_cols.values, handle_missing='return_nan',return_df= True)
X_train = enc.fit_transform(X_train)
X_test = enc.transform(X_test)

# check the data types now 
X_train.info(verbose=True, null_counts=True)
```

```python
# check if there is any constant in the dataset

from sklearn.feature_selection import VarianceThreshold

# setting threshold to zero
var_thres=VarianceThreshold(threshold=0)
var_thres.fit(X_train)
len(X_train.columns[var_thres.get_support()])

constant_columns = [column for column in X_train.columns
                    if column not in X_train.columns[var_thres.get_support()]]

# remove constant from the dataset
X_train = X_train.drop(constant_columns,axis=1)
X_test = X_test.drop(constant_columns,axis=1)

# print the name of the constant deleted
for column in constant_columns:
    print(column)
```

```python
# check the data shape now 
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# list of columns which have more than 50% missing data
percent_missing = X_train.isnull().sum() * 100 / len(data)
missing_value_df = pd.DataFrame({'column_name': X_train.columns,
                                 'percent_missing': percent_missing})
missing_values = missing_value_df.loc[(missing_value_df.percent_missing>1),'column_name'].values
print(missing_values)
```

```python
# fill the missing value using KNNImputer 

#import the relevant methods
from sklearn.impute import KNNImputer
nan = np.nan

# save the column names first
X_train_col = X_train.columns
X_test_col = X_test.columns

#Apply KNN imputer
imputer = KNNImputer(n_neighbors=2, weights="uniform")
imputed_X_train = imputer.fit_transform(X_train)
imputed_X_test = imputer.transform(X_test)

# Convert output to the data frames
X_train = pd.DataFrame.from_records(imputed_X_train)
X_train.columns = X_train_col

X_test = pd.DataFrame.from_records(imputed_X_test)
X_test.columns = X_test_col

# check the data types now, there is no null values
X_train.info(verbose=True, null_counts=True)
```

```python
# Feature Selection based on Mutual Information

from sklearn.feature_selection import mutual_info_regression

# determine the mutual information
mutual_info = mutual_info_regression(X_train, y_train)
mutual_info = pd.Series(mutual_info)
mutual_info.index = X_train.columns
mutual_info.sort_values(ascending=False).head(50)
```

```python
# ploting the ordered mutual_info values per feature
mutual_info.head(29).sort_values(ascending=False).plot.bar(figsize=(20, 8))
```

```python
# we will keep only the top 26 important features

from sklearn.feature_selection import SelectKBest

sel_cols = SelectKBest(mutual_info_regression, k=26)
sel_cols.fit(X_train, y_train)
selected_cols = X_train.columns[sel_cols.get_support()]

X_train = X_train.loc[:, selected_cols]
X_test = X_test.loc[:, selected_cols]
```

```python
# check the data shape now 
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# check the correlated features using Pearson Correlation

corrmat = X_train.corr()
fig, ax = plt.subplots()
fig.set_size_inches(11,11)
sns.heatmap(corrmat)
```

```python
# this function will select highly correlated features

def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
# number of features which are more than 80% correlated
corr_features = correlation(X_train, 0.8)
len(set(corr_features))
```

```python
# list of correlated columns
corr_features
```

```python
# removing correlated columns from the dataset

X_train = X_train.drop(corr_features,axis=1)
X_test = X_test.drop(corr_features,axis=1)

# check the data shape now 
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```
